## Função Step

Usado pelos primeiros neuronios artificiais criados, no Modelo de McCulloch e Pitts (1943), onde eles computavam uma soma ponderada e aplicavam a função, e no Perceptron (1958, Frank Rosenblatt), que seguia a mesma lógica, mas, dessa vez, atualizava os pesos com algoritmos supervisionados.

### Importância

A função Step é extremamente util para problemas linearmente separáveis.

### Limitações e declínio

Mas, atualmente caiu em desuso, por não ser diferenciável. Em outras palavras, não é possível aplicar tecnicas de otimização mais avançadas, como o gradiente descente.

Entretanto, isso não tira sua importancia nos dias de hoje e para a história da Inteligência Artificial.

### Função matemática:
$$
f(x) = \left\{
  \begin{array}{l}
  0,\, se\; x < 1 \\
  1,\, se\; x \ge 1
  \end{array}
  \right.
$$

### Codificação da função:

```python
def Step(x):
  if x >= 1:
    return 1
  else:
    return 0
```

```python
print(Step(0.5))
print(Step(1))
print(Step(1.5))
```

## Sigmoid

Sua origem matemática se dá pelo estudo de **biologia** e **estatística**, em especial na modelagem do crescimento populacional.

Por volta dos anos 1980, a função começou a se popularizar no ramo de Redes Neurais. Assim, substituindo a função Step usada na década de 1960

### Importância

O que a fez se popularizar foi:

- Ela ser diferenciável, podendo utilizar o algoritmo de backpropagation.
- Introduz a não-linearidade nos modelos, fazendo com que as redes neurais possam aprender padrões mais complexos.
- O valor de saida é de **0** a **1**, o que ajuda a modelar probabilidades e facilita a interpretação.

### Limitações e declínio

Ela acabou perdendo espaço para outras funções. Por conta de:

- O desvanecimento do gradiente, pois, para valores extremos, o gradiente da função se aproxima a 0, dificultando a aprendizagem da rede.
- Computacionalmente mais custosa em relação às funções mais modernas.

### Uso atual

Atualmente é muito utilizada em classificações binárias, pois a saida entre 0 e 1 é útil para representar uma probabilidade.

### Função matemática
$$
f(x) = \frac{1}{1 + e^{-x}}
$$

### Codificando a função:

```python
def Sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
print(Sigmoid(-5))
print(Sigmoid(0))
print(Sigmoid(10))
```

## Tangente Hiperbólica


Essa função tem sua origem na matemática pura, uma função hiperbólica análoga à tangente trigonométrica. A sua forma simétrica e suavemente crescente é muito útil em diversas áreas, inclusive em redes neurais.

A tanh começou a ser usada na decada de 1980, junto com a popularização da função sigmoid, impulsionada pelo algoritmo de backpropagation.

### Importância

- Assim como a sigmoid, ela é **diferencial** e **não linear**.
- A sua saida é entre **-1** e **1**, ou seja, é centrada em zero, o que facilita o treinamento dos modelos.
- Tem gradientes maiores, o que reduz o problema do desvanimento do gradiente.

### Limitações e declínio

- A função sofre o problema de saturação para entradas muito positivas ou muito negativas, causando um desvanimento do gradiente.
- Alto custo computacional em relação às funções mais modernas.  

### Uso atual

- Ainda usada em redes neurais mais intermediárias.
- Presente em redes neurais recorrentes, por conta de sua suavidade e estabilidade numérica.

### Função matemática
$$
f(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

### Codificando a função:

```python
def Tanh(x):
  nominador = np.exp(x) - np.exp(-x)
  denominador = np.exp(x) + np.exp(-x)
  return nominador / denominador
```

```python
print(Tanh(-5))
print(Tanh(0))
print(Tanh(3))
```

## ReLU - Rectified Linear Units

É uma função simples, parcialmente linear: onde se o valor de entrada for positivo, retorna o próprio valor, se for negativo, retorna 0.

Ela foi formulada na década de 1980, junto com as outras funções. Mas só foi popularizada a partir de 2010, quando alguns pesquisadores descobriram que essa função resolvia problemas críticos enfrentados pela Sigmoid e Tanh.

### Importância
Por que a ReLU foi revolucionária:

- Computacionalmente simples, pois não envolve calculos exponenciais, mas sim comparações.
- Evita a saturação, pois, para valores positivos, o gradiente é constante (1), o que mitiga o problema do desvanecimento do gradiente.
- Por ser computacionalmente simples, possibilitou treinar redes com muitas camadas, o que impulsionou o **deep learning moderno**.

### Limitações

- "Morte de neurônios": se o valor de entrada de um neurônio for sempre negativo, a saida sempre será 0, e ele para de "aprender".
  - Para mitigar este problema, foram criadas variações desta função.
- É sensivel à **inicialização dos pesos** e à **taxa de aprendizado**.

### Uso atual

- É a função **mais usada** em redes neurais modernas.
- Presente em diversas arquiteturas, como as **CNNs, ResNets, Transformers** e etc.

### Função matemática
$$
f(x) = \left\{
  \begin{array}{l}
  0,\, se\; x < 0 \\
  x,\, se\; x \ge 0
  \end{array}
  \right.
$$

### Codificando a função:

```python
def ReLU(x):
  if x < 0:
    return 0
  else:
    return x
```

```python
print(ReLU(-3))
print(ReLU(1))
print(ReLU(3))
```

## Linear

A função linear é a forma mais simples de função, sua saída é diretamente proporcional à entrada. É usada em diversas áreas da matemática e tem aplicações centenárias.

Esta função foi considerada pelos pesquisadores quando buscavam funções continuas, pois estavam procurando aproximações mais suaves e diferenciáveis.

### Importância

- Simples e eficiente, é facil de computar e interpretar.
- Usada para modelos de regressão, especialmente para previsões contínuas.
- Serve como um "pass-through", pois apenas multiplica e soma os pesos, não alterando o valor final.

### Limitações

Por mais ser simples, ela tem uma **limitação crítica** nas redes neurais.

- Não permite aprendizado complexo, pois, quando todas as camadas são lineares, a rede inteira **colapsa** em uma única transformação linear. Ou seja, não há ganho de complexidade com camadas adicionais.

### Uso Atual

A função é usada em principalmente em situações específicas.

- Usada como camada de saída de redes de regressão.
- Aquiteturas específicas que combinam saídas lineares com as funções de perda adequadas, como o erro quadrático médio.
- Usado também como bloco intermediário em algumas arquiteturas avançadas, antes de alguma normalização ou camada de ativação específica.

### Função matemátia
$$
f(x) = x
$$

### Codificando a função

```python
def Linear(x):
  return x
```

```python
print(Linear(-5))
print(Linear(0))
print(Linear(10))
```

## Softmax

A função Softmax é uma generalização da função sigmoide para múltiplas classes. Ela transforma um vetor de valores reais em um vetor de probabilidades que somam 1. Essa função tem origem na **estatística** e na **teoria da decisão**, especialmente na modelagem de distribuições de probabilidade sobre classes exclusivas.

No contexto das inteligencia computacional, foi adotada principalmente para redes multiclasses de classificação e se tornou o padrão em problemas onde a saída precisa representar uma distribuição de probabilidade sobre várias classes.

### Importância

- Transforma scores brutos em **valores probabilisticos interpretaveis**.
- A saída é normalizada com a soma 1, o que possibilita usar funções de erro baseadas em probabilidades, como o **cross-entropy**.

### Limitações

- Pode ser sensível para valores extremamente altos, já que usam calculos exponencias.
- Em tarefas de multiplas etiquetas, ela não é adequada, pois presume que apenas uma é correta.

### Uso atual

- Presente em quase todos os **classificadores multiclasse** baseados em redes neurais.
- Muito presente em outras arquiteturas, como **CNNs**, **Tranformers**, **RNNs** e entre outras.

### Função matemática
$$
Y = \frac{e(x)}{\sum{e(x)}}
$$

### Codificando a função

```python
# Essa função recebe np.array's
def Softmax(x):
  e_x = np.exp(x)
  return e_x / np.sum(e_x)
```

```python
x = np.array([1.0, 2.0, 0.1])
print(Softmax(x))
```
